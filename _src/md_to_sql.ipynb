````python
import os, re
````

````python
f_out = open('_md_split/all_queries.sql', 'w')
for filename in os.listdir('./_md'):
    base_name, extension = os.path.splitext(filename)
    if extension != '.md':
        continue
    
    try:
        os.mkdir(base_name)
    except Exception as e:
        print(e)
        
    print(filename)
    with open(filename, 'r') as f_in:
        is_in_codeblock = False
        f_md_out = None
        for line in f_in:
            line = line.replace(' '*4, '\t')
            line = line.strip('\n\r ')        

            if re.match(r'[A-Z]{1,3}\d{2}:', line):
                if f_md_out is not None:
                    f_md_out.close()
                filename = os.path.join('_md_split', base_name, line.replace(' ', '_').replace('/', '_').replace(':', '').replace('?', '').strip('.') + '.md')
                f_md_out = open(filename, 'w')
                print(line)
                
            if f_md_out is None:
                continue
            
            if len(line) > 0 and line[0] == '\t':
                if not is_in_codeblock:
                    f_md_out.write('\n```sql\n')
                    is_in_codeblock = True
                line = line[1:].replace('\t', ' '*4)
                f_out.write(line)
                f_out.write('\n')
            elif not line and is_in_codeblock:
                continue
            elif is_in_codeblock:
                f_md_out.write('```\n')
                f_out.write('\n')
                is_in_codeblock = False
            
            f_md_out.write(line)
            f_md_out.write('\n')
        if f_md_out is not None: 
            f_md_out.close()    

f_out.close()
````
